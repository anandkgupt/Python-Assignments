5. What are iterators in Python and how do they differ from iterables?

Ans- Iterators in Python:
An iterator is an object in Python that enables traversal (iteration) through a sequence of elements, one element at a time. It doesnâ€™t require the entire sequence to be stored in memory, which makes it memory-efficient.

Iterables in Python:
An iterable is an object that can be converted into an iterator. It is any object that implements the __iter__() method, which returns an iterator.

Difference Between an Iterable and an Iterator:

Iterable: Any object that can return an iterator (e.g., lists, strings, tuples). It must implement the __iter__() method.

Example: for loops work on iterables because they implicitly create iterators.

Iterator: An object with a state that remembers where it is during traversal. It must implement both the __iter__() and __next__() methods. Iterators exhaust their elements after traversal, unlike iterables.

```python
## Example of Iterating over a list using an iterator
my_list = [10, 20, 30]
iterator = iter(my_list)  # Create an iterator object from the list

print(next(iterator))
print(next(iterator))
print(next(iterator))
```

6. Explain the concept of generators in Python and how they are defined.

Ans- Generators in Python:
Generators are a special type of iterable in Python that allow you to iterate through a sequence of values lazily, one at a time, without storing the entire sequence in memory. They are used to produce values on demand, making them highly memory-efficient, especially for large datasets or infinite sequences.

Defining a Generator:
A generator is defined like a normal function but uses yield instead of return.

```python
## Example
def count_up_to(max_value):
    current = 1
    while current <= max_value:
        yield current  # Produce the current value
        current += 1
# Using the generator
gen = count_up_to(3)
print(next(gen))
print(next(gen))
print(next(gen))
```

7. What are the advantages of using generators over regular functions?

Advantages of Using Generators Over Regular Functions:
Generators offer significant benefits over regular functions, particularly in scenarios where efficiency and performance are critical.

(1) Memory Efficiency:

Generators produce values one at a time using the yield statement, avoiding the need to store the entire sequence in memory.

Regular Functions, on the other hand, often return large data structures like lists, which can consume a significant amount of memory.

(2) Lazy Evaluation:

Generators compute values on demand, i.e., only when next() is called or when the generator is iterated.

This is useful when processing a large dataset or an infinite sequence where you may not need all values at once.

(3) Simplified Code:

Generators simplify the implementation of iterators and stateful functions by maintaining their own state.

Regular functions require more boilerplate code to achieve the same functionality.

```python
## Example of Using a generator to produce the first n numbers
def number_generator(n):
    for i in range(n):
        yield i

# No memory overhead for a large sequence
gen = number_generator(10**6)
print(next(gen))
```

```python
## Example of Using a regular function for the same task
def number_list(n):
    return [i for i in range(n)]

# Creates a large list in memory
lst = number_list(10**6)
print(lst[0])
```

9. Explain the purpose and usage of the `map()` function in Python.

Ans- map() Function in Python:

The map() function in Python applies a given function to each item of an iterable (like a list, tuple, or string) and returns an iterator that produces the results. It is commonly used for transforming data in a functional programming style.

Purpose of map():

The main purpose of map() is to simplify the process of applying a function to each element of an iterable without explicitly writing a loop. It improves code readability and efficiency, especially for large datasets.

Use Cases of map():

Data Transformation: Quickly modify each element in a dataset (e.g., converting units, cleaning text).

Parallel Processing: Pair elements from multiple iterables for simultaneous computation.

Data Pipelines: Integrate into functional-style pipelines for cleaner code.





```python
## example of using map() to convert all strings in a list to uppercase:
words = ['apple', 'banana', 'cherry']
uppercased = map(str.upper, words)
print(list(uppercased))
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans- Difference Between map(), reduce(), and filter() Functions in Python:

The map(), reduce(), and filter() functions are functional programming tools in Python, often used for processing and transforming iterables. While they have similar applications, they serve distinct purposes.

(1) map() Function:

Purpose: Applies a given function to each item in an iterable and returns an iterator with the transformed values.

Returns: An iterator with the same number of elements as the input iterable.

(2) filter() Function:

Purpose: Filters elements from an iterable based on a function that returns True or False.

Returns: An iterator containing only the elements for which the function evaluates to True.

(3) reduce() Function:

Purpose: Reduces an iterable to a single cumulative value by repeatedly applying a function.

Returns: A single value.

```python
## example of map() function
numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))
```

```python
## example of filter() function
numbers = [1, 2, 3, 4]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))
```

```python
## example of reduce() function
from functools import reduce
numbers = [1, 2, 3, 4]
sum_result = reduce(lambda x, y: x + y, numbers)
print(sum_result)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(num):
  """Checks if a given number is prime.

  Args:
    num: The number to check.

  Returns:
    True if the number is prime, False otherwise.
  """

  if num <= 1:
    return False
  if num <= 3:
    return True
  if num % 2 == 0 or num % 3 == 0:
    return False

  i = 5
  while i * i <= num:
    if num % i == 0 or num % (i + 2) == 0:
      return False
    i += 6

  return True

# Test the function for numbers from 1 to 200
for num in range(1, 201):
  if is_prime(num):
    print(num, "is prime")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration

        self.count += 1
        self.a, self.b = self.b, self.a + self.b
        return self.a - self.b

# Example usage:
for num in FibonacciIterator(10):
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):
  """Generates powers of 2 up to a given exponent.

  Args:
    exponent: The maximum exponent.

  Yields:
    The next power of 2.
  """

  power = 1
  for i in range(exponent + 1):
    yield power
    power *= 2

# Example usage:
for power in powers_of_two(5):
  print(power)
```

11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- â‚¬ if the value of the
order is smaller than 100,00 â‚¬.

Write a Python program using lambda and map.

```python
def calculate_order_values(orders):
    """
    Processes a list of orders and returns a list of 2-tuples.
    Each tuple contains the order number and the adjusted total price.

    Args:
        orders (list): A list of sublists, where each sublist contains:
                       [order_number, price_per_item, quantity].

    Returns:
        list: A list of 2-tuples (order_number, total_price).
    """
    return list(map(
        lambda order: (
            order[0],
            order[1] * order[2] + (10 if order[1] * order[2] < 100 else 0)
        ),
        orders
    ))

# Example orders list
orders = [
    [34587, 40.95, 4],
    [98762, 56.80, 5],
    [77226, 32.95, 3],
    [88112, 24.99, 3]
]

# Calculate and print the results
results = calculate_order_values(orders)
print(results)
```
