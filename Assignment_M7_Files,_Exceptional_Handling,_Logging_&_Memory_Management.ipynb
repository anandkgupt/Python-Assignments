Q1. What is the difference between interpreted and compiled languages?

Ans: The primary difference between interpreted and compiled languages lies in how the code is executed on the computer.

(1) Interpreted Languages:

Execution: The code is executed line-by-line or statement-by-statement by an interpreter during runtime.

Translation: No separate compilation step; the source code remains in its original format.

Speed: Slower execution compared to compiled languages because translation happens on the fly.

Error Detection: Errors are detected at runtime, making debugging easier but delaying the discovery of issues.

Examples: Python, JavaScript, Ruby, PHP

(2) Compiled Languages:

Execution: The code is first translated (compiled) into machine code or an intermediate language by a compiler before execution.

Translation: Once compiled, the executable file can run without the source code or the compiler.

Speed: Faster execution because the code is already translated into machine language.

Error Detection: Errors are detected at compile-time, making debugging more upfront but often quicker to resolve in the long term.

Examples: C, C++, Rust, Go.

Ans: Difference Between os.fork() and multiprocessing in Python:

Both os.fork() and the multiprocessing module in Python are used to create new processes. However, they have significant differences in their functionality, use cases, and behavior. Here's a comparison of the two:

1.os.fork():

The os.fork() function is a low-level method that creates a child process by duplicating the parent process. It creates a new process by splitting the execution flow into two: one in the parent process and one in the child process.

The child process gets a copy of the parent's memory and environment, but the process IDs are different.

2.multiprocessing Module:

The multiprocessing module is a higher-level abstraction that provides a more powerful and flexible way to create and manage processes.

It allows you to spawn multiple processes, synchronize them, and communicate between them using shared memory, queues, and pipes.

Unlike os.fork(), multiprocessing works on both Unix and Windows systems.

Q4. Write a Python script that reads from one file and writes its content to another file.

```python
# File paths
source_file = "source.txt"
destination_file = "destination.txt"

try:
    # Open the source file for reading
    with open(source_file, "r") as src:
        # Read the contents of the source file
        content = src.read()

    # Open the destination file for writing
    with open(destination_file, "w") as dest:
        # Write the content to the destination file
        dest.write(content)

    print(f"Contents of '{source_file}' have been successfully written to '{destination_file}'.")
except FileNotFoundError:
    print(f"Error: The file '{source_file}' does not exist.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

Q18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?

```python
import logging
from logging.handlers import RotatingFileHandler

# Define the logging setup
def setup_logger():
    # Create a logger
    logger = logging.getLogger('MyLogger')
    logger.setLevel(logging.DEBUG)  # Set the logging level to DEBUG

    # Create a rotating file handler that will rotate the log file after 1MB (1 * 1024 * 1024 bytes)
    handler = RotatingFileHandler('app.log', maxBytes=1 * 1024 * 1024, backupCount=3)

    # Create a log formatter
    formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
    handler.setFormatter(formatter)

    # Add the handler to the logger
    logger.addHandler(handler)

    return logger

# Example usage of the logger
if __name__ == "__main__":
    logger = setup_logger()

    # Log some messages
    for i in range(1000):
        logger.debug(f"This is debug message {i}")
        logger.info(f"This is info message {i}")
        logger.warning(f"This is warning message {i}")
        logger.error(f"This is error message {i}")
        logger.critical(f"This is critical message {i}")
```

Q21. Write a Python program that reads a file and prints the number of occurrences of a specific word.

```python
# Program to count occurrences of a specific word in a file

def count_word_in_file(file_name, word):
    try:
        # Open the file using a context manager
        with open(file_name, 'r') as file:
            # Initialize a counter for the word
            word_count = 0

            # Read each line in the file
            for line in file:
                # Split the line into words and count the occurrences of the word
                word_count += line.lower().split().count(word.lower())

        return word_count

    except FileNotFoundError:
        print(f"The file '{file_name}' was not found.")
        return 0

# Example usage
file_name = 'example.txt'  # Replace with the path to your file
word = 'python'  # Replace with the word you want to count

count = count_word_in_file(file_name, word)
print(f"The word '{word}' occurred {count} times in the file '{file_name}'.")
```

```python
import os

def check_file_empty(file_name):
    # Check if the file exists and is empty
    if os.path.exists(file_name) and os.stat(file_name).st_size == 0:
        print(f"The file '{file_name}' is empty.")
        return True
    elif not os.path.exists(file_name):
        print(f"The file '{file_name}' does not exist.")
        return False
    else:
        print(f"The file '{file_name}' is not empty.")
        return False

# Example usage
file_name = 'example.txt'  # Replace with the path to your file
if not check_file_empty(file_name):
    with open(file_name, 'r') as file:
        content = file.read()
        print(content)
```

Q23. Write a Python program that writes to a log file when an error occurs during file handling.

```python
import logging

# Configure logging to write errors to a log file
logging.basicConfig(
    filename='file_handling_errors.log',  # Log file to store errors
    level=logging.ERROR,                  # Log only errors
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def handle_file_operations(file_name):
    try:
        # Trying to open a file that may not exist or have permission issues
        with open(file_name, 'r') as file:
            content = file.read()
            print(content)

    except FileNotFoundError as fnf_error:
        # Log the error if the file is not found
        logging.error(f"FileNotFoundError: {fnf_error}. The file '{file_name}' was not found.")
        print(f"Error: The file '{file_name}' was not found. Please check the file path.")

    except PermissionError as perm_error:
        # Log the error if there are permission issues
        logging.error(f"PermissionError: {perm_error}. You do not have permission to access the file '{file_name}'.")
        print(f"Error: You do not have permission to access the file '{file_name}'.")

    except Exception as error:
        # Log any other generic errors
        logging.error(f"Unexpected error: {error}. An error occurred while handling the file '{file_name}'.")
        print("An unexpected error occurred. Please check the log for details.")

# Example usage
file_name = 'non_existent_file.txt'  # You can change this to a file that may cause an error
handle_file_operations(file_name)
```
