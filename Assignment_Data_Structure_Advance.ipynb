```python
# 1. Reverse a string
def reverse_string(s):
    return s[::-1]
# Usage: reverse_string("hello")

# 2. Count the number of vowels in a string
def count_vowels(s):
    vowels = 'aeiouAEIOU'
    return sum(1 for char in s if char in vowels)
# Usage: count_vowels("hello")

# 3. Check if a given string is a palindrome
def is_palindrome(s):
    return s == s[::-1]
# Usage: is_palindrome("madam")

# 4. Check if two strings are anagrams
from collections import Counter
def are_anagrams(s1, s2):
    return Counter(s1) == Counter(s2)
# Usage: are_anagrams("listen", "silent")

# 5. Find all occurrences of a substring within another string
def find_substring_occurrences(s, sub):
    return [i for i in range(len(s) - len(sub) + 1) if s[i:i + len(sub)] == sub]
# Usage: find_substring_occurrences("hello hello", "lo")

# 6. Basic string compression using counts of repeated characters
def compress_string(s):
    if not s:
        return ""
    compressed, count = [], 1
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1
    compressed.append(s[-1] + str(count))
    return ''.join(compressed)
# Usage: compress_string("aabcccccaaa")

# 7. Determine if a string has all unique characters
def all_unique_characters(s):
    return len(set(s)) == len(s)
# Usage: all_unique_characters("abcdef")

# 8. Convert a string to uppercase or lowercase
def to_uppercase(s):
    return s.upper()
def to_lowercase(s):
    return s.lower()
# Usage: to_uppercase("hello"), to_lowercase("HELLO")

# 9. Count the number of words in a string
def count_words(s):
    return len(s.split())
# Usage: count_words("Hello world")

# 10. Concatenate two strings without using the + operator
def concatenate_strings(s1, s2):
    return "{}{}".format(s1, s2)
# Usage: concatenate_strings("Hello", "World")

# 11. Remove all occurrences of a specific element from a list
def remove_all_occurrences(lst, element):
    return [x for x in lst if x != element]
# Usage: remove_all_occurrences([1, 2, 3, 2, 4], 2)

# 12. Find the second largest number in a list
def second_largest(lst):
    unique_numbers = list(set(lst))
    unique_numbers.sort()
    return unique_numbers[-2] if len(unique_numbers) >= 2 else None
# Usage: second_largest([1, 2, 3, 4, 5])

# 13. Count occurrences of each element in a list and return a dictionary
def element_counts(lst):
    return {x: lst.count(x) for x in set(lst)}
# Usage: element_counts([1, 2, 2, 3, 3, 3])

# 14. Reverse a list in-place without using built-in functions
def reverse_list(lst):
    for i in range(len(lst) // 2):
        lst[i], lst[-i - 1] = lst[-i - 1], lst[i]
    return lst
# Usage: reverse_list([1, 2, 3, 4])

# 15. Remove duplicates from a list while preserving order
def remove_duplicates(lst):
    seen = set()
    return [x for x in lst if not (x in seen or seen.add(x))]
# Usage: remove_duplicates([1, 2, 2, 3, 3, 4])

# 16. Check if a list is sorted
def is_sorted(lst):
    return lst == sorted(lst) or lst == sorted(lst, reverse=True)
# Usage: is_sorted([1, 2, 3])

# 17. Merge two sorted lists
def merge_sorted_lists(lst1, lst2):
    return sorted(lst1 + lst2)
# Usage: merge_sorted_lists([1, 3, 5], [2, 4, 6])

# 18. Find the intersection of two lists
def list_intersection(lst1, lst2):
    return list(set(lst1) & set(lst2))
# Usage: list_intersection([1, 2, 3], [2, 3, 4])

# 19. Find the union of two lists without duplicates
def list_union(lst1, lst2):
    return list(set(lst1) | set(lst2))
# Usage: list_union([1, 2, 3], [3, 4, 5])

# 20. Shuffle a list without using built-in functions
import random
def shuffle_list(lst):
    for i in range(len(lst)):
        swap_index = random.randint(0, len(lst) - 1)
        lst[i], lst[swap_index] = lst[swap_index], lst[i]
    return lst
# Usage: shuffle_list([1, 2, 3, 4, 5])

# 21. Find common elements between two tuples
def tuple_intersection(t1, t2):
    return tuple(set(t1) & set(t2))
# Usage: tuple_intersection((1, 2, 3), (2, 3, 4))

# 22. Find intersection of two sets input by the user
def user_set_intersection():
    set1 = set(map(int, input("Enter integers for first set: ").split(',')))
    set2 = set(map(int, input("Enter integers for second set: ").split(',')))
    return set1 & set2

# 23. Concatenate two tuples
def concatenate_tuples(t1, t2):
    return t1 + t2
# Usage: concatenate_tuples((1, 2), (3, 4))

# 24. Elements in first set but not in the second
def set_difference(set1, set2):
    return set1 - set2
# Usage: set_difference({1, 2, 3}, {2, 3, 4})

# 25. Return elements of a tuple within a specified range of indices
def slice_tuple(t, start, end):
    return t[start:end]
# Usage: slice_tuple((1, 2, 3, 4, 5), 1, 3)

# 26. Union of two sets
def set_union(set1, set2):
    return set1 | set2
# Usage: set_union({1, 2, 3}, {3, 4, 5})

# 27. Find max and min in a tuple using tuple unpacking
def tuple_min_max(t):
    return min(t), max(t)
# Usage: tuple_min_max((1, 2, 3, 4, 5))

# 28. Union, intersection, and difference of two sets
def set_operations(set1, set2):
    return set1 | set2, set1 & set2, set1 - set2
# Usage: set_operations({1, 2, 3}, {2, 3, 4})

# 29. Count occurrences of an element in a tuple
def count_in_tuple(t, element):
    return t.count(element)
# Usage: count_in_tuple((1, 2, 2, 3), 2)

# 30. Symmetric difference of two sets
def symmetric_difference(set1, set2):
    return set1 ^ set2
# Usage: symmetric_difference({1, 2, 3}, {2, 3, 4})

# 31. Word frequency dictionary from a list of words
def word_frequencies(words):
    return {word: words.count(word) for word in set(words)}
# Usage: word_frequencies(["apple", "banana", "apple"])

# 32. Merge two dictionaries with added values for common keys
def merge_dictionaries(d1, d2):
    result = d1.copy()
    for k, v in d2.items():
        result[k] = result.get(k, 0) + v
    return result
# Usage: merge_dictionaries({'a': 1, 'b': 2}, {'b': 3, 'c': 4})

# 33. Access a value in a nested dictionary
def nested_dict_access(d, keys):
    for key in keys:
        if isinstance(d, dict):
            d = d.get(key, None)
        else:
            return None
    return d
# Usage: nested_dict_access({'a': {'b': {'c': 1}}}, ['a', 'b', 'c'])

# 34. Sort a dictionary by its values
def sort_dict_by_values(d, ascending=True):
    return dict(sorted(d.items(), key=lambda item: item[1], reverse=not ascending))
# Usage: sort_dict_by_values({'a': 3, 'b': 1, 'c': 2})

# 35. Invert a dictionary, handling duplicate values
def invert_dictionary(d):
    inverted = {}
    for k, v in d.items():
```
